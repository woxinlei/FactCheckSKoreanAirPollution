```python
#%%
import numpy as np

def create_sequence_data_for_points(grouped, time_steps, features, target_column='pm2_5'):
    """
    针对每个 (point_x, point_y) 来创建时间序列数据。
    
    参数:
    - grouped: 按照 (point_x, point_y) 分组的数据
    - time_steps: 每个时间序列的时间步长
    - features: 用于训练的特征列
    - target_column: 目标列，默认是 'pm2_5'
    
    返回:
    - X_seq_dict: 每个 (point_x, point_y) 对应的输入特征时间序列数据
    - y_seq_dict: 每个 (point_x, point_y) 对应的标签时间序列数据
    """
    X_seq_dict = {}
    y_seq_dict = {}
    
    for (point_x, point_y), group in grouped:
        # 获取该点的数据，取出特征和目标列
        group_data = group[features]  # 获取所有特征
        target_data = group[target_column]  # 获取目标值，即 pm2_5
        
        # 转换为 NumPy 数组
        X = group_data.values
        y = target_data.values
        
        # 使用 create_sequence_data 来生成时间序列数据
        X_seq, y_seq = create_sequence_data(X, y, time_steps)
        
        # 存储该点的时间序列数据
        X_seq_dict[(point_x, point_y)] = X_seq
        y_seq_dict[(point_x, point_y)] = y_seq
    
    return X_seq_dict, y_seq_dict


def create_sequence_data(X, y, time_steps):
    """
    将数据转化为时间序列格式。
    
    参数:
    - X: 特征数据
    - y: 目标值数据
    - time_steps: 每个时间序列的步长
    
    返回:
    - X_seq: 生成的时间序列特征数据
    - y_seq: 生成的时间序列目标数据
    """
    X_seq = []
    y_seq = []
    
    for i in range(len(X) - time_steps):
        X_seq.append(X[i:i + time_steps])  # 取时间步数长度的数据
        y_seq.append(y[i + time_steps])    # 对应的标签（下一时刻的 pm2_5）
    
    return np.array(X_seq), np.array(y_seq)
```
